```python
import math
import matplotlib.pyplot as plt
def mylinspace(start, stop, num):
 """Generate a list of evenly spaced values between start and stop."""
 step = (stop - start) / (num - 1)
 return [start + i * step for i in range(num)]
def mynormal_pdf(x, mean, std_dev):
 """Calculate the PDF of a normal distribution."""
 coefficient = 1 / (math.sqrt(2 * math.pi) * std_dev)
 exponent = math.exp(-((x - mean) ** 2) / (2 * std_dev ** 2))
 return coefficient * exponent


```

```python

# Generate data for x values
x = mylinspace(-10, 20, 100)
# Parameters
mean, std_dev = 4.9, 8.7
# Generate PDF values
normal_dist = [mynormal_pdf(val, mean, std_dev) for val in x]
# Plot PDF for normal distribution
plt.figure(figsize=(12, 6))
plt.plot(x, normal_dist, label=f'Normal (µ={mean}, σ={std_dev})', color='blue')
# Add labels and title
plt.title("Characterization of Density Functions")
plt.xlabel("X values")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
# Show the plot
plt.show()
```

```python
#Write a Python program to characterize the density functions.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
# Plot PDF for normal distribution
x = np.linspace(-10, 20, 100)
pdf = norm.pdf(x, loc=4.9, scale=8.7)
#The location (loc) keyword specifies the mean. The scale (scale) keyword specifies the standard deviation.

plt.plot(x, pdf, label='Normal PDF')
plt.title('Probability Density Function (Normal)')
plt.xlabel('x')
plt.ylabel('Density')
plt.legend()
plt.show()
```
