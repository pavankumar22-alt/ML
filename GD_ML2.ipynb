```python
import numpy as np
```

```python
X=np.asarray([20,15,10,13,12])
Y=np.asarray([44,22,25,32,27])
```

```python
b1=0
b0=0
epochs=1000
L=0.001
n=np.size(X)
```

```python
# Performing Gradient Descent
for i in range(epochs):
 db1, db0, err, Total_Err_x, Total_Err_y = 0, 0, 0, 0, 0
 for x,y in zip(X,Y):
  y_pred = b1*x + b0 # The current predicted value of y
  err=(y - y_pred)
  Total_Err_x = Total_Err_x+ x*(err)
  Total_Err_y = Total_Err_y + err
 D_b1 = (-2/n) * Total_Err_x # Derivative wrt b1
 D_b0 = (-2/n) * Total_Err_y # Derivative wrt b0
 b1 = b1 -L * D_b1 # Update b1
 b0 = b0 - L * D_b0 # Update b0
print(b1,b0)
```

```python
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
Y_pre = b1 * X + b0
print(mean_squared_error(Y, Y_pre)) # Use the original Y array
coeffiecnt_of_determiATION = 1 - (sum((Y - Y_pre)**2) / sum((Y - np.mean(Y))**2)) # Use the original Y array
print(coeffiecnt_of_determiATION)
plt.figure(figsize=(10, 5))
plt.scatter(X, Y, color='red')
plt.plot(X, Y_pre, color='blue')
plt.show()
```
