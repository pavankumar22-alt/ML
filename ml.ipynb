```python
import pandas as pd
import numpy as np
x=np.asarray([2,3,4,5,6])
y=np.asarray([7,9,10,14,15])
```

```python
df=pd.DataFrame(
    {
        'X':x,
        'Y':y }
)
df.head()
```

```python
xmean=np.mean(x)
ymean=np.mean(y)
xy=x*y
xymean=np.mean(xy)
x2=x**2
x2mean=np.mean(x2)
meansq=xmean**2
b1=(xymean-(xmean*ymean))/(x2mean-meansq)
b0=ymean-b1*xmean
print(b1,b0)
```

```python
ypred=b1*x+b0
ydif=y-ypred
mse=np.sum(ydif**2)/len(y)
print(mse)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.plot(x,ypred)
plt.show()
```

```python
ssre=(y-ypred)**2
stotal=(y-ymean)**2
rs2=1-(np.sum(ssre)/np.sum(stotal))
print(rs2)
```
