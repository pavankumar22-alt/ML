```python
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
import pandas as pd

# Sample data
data = {
    'X': [1, 2, 3, 4, 5],
    'y': [2, 4, 5, 4, 6]
}
df = pd.DataFrame(data)

# Initialize and fit linear regression model
model1 = smf.ols('y ~ X', data=df)
result = model1.fit()

# Print model summary
print(result.summary())

# Predict y values
ypred = result.predict(df['X'])
print("Predicted values:\n", ypred)

# Print model parameters
print("\nCoefficients:\n", result.params)
print("\nP-values:\n", result.pvalues)
print("\nR-squared:\n", result.rsquared)

# Plot regression against actual data
plt.figure(figsize=(10, 5))
plt.plot(df['X'], df['y'], 'o', label='Actual data')  # Scatter plot
plt.plot(df['X'], ypred, 'r', linewidth=2, label='Regression line')  # Regression line
plt.xlabel('X values')
plt.ylabel('y values')
plt.title('Linear Regression using `statsmodels`')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Data
X = np.asarray([20, 5, 10, 13, 12])
y = np.asarray([44, 22, 25, 32, 27])

# Before reshaping
print("Before reshaping")
print(X)

# Reshape X to a 2D array
X = X.reshape(-1, 1)

# After reshaping
print("After reshaping")
print(X)

# Print X and y values
print("X Values:\n", X)
print("Y Values:\n", y)

# Create and fit linear regression model
model = LinearRegression()
model.fit(X, y)

# Print model coefficients
print("b1 (Slope):\n", model.coef_)
print("b0 (Intercept):\n", model.intercept_)

# Predict using the training set
yhat = model.predict(X)
print("Predictions:\n", yhat)

# Model performance
r_sq = model.score(X, y)
print("Coefficient of Determination (R^2):", r_sq)
print("MSE:", mean_squared_error(y, yhat))
print("RMSE:", np.sqrt(mean_squared_error(y, yhat)))
```

```python
plt.figure(figsize=(10,5))
plt.plot(X, y, 'o', label='Actual data')
plt.plot(X, yhat, 'r', linewidth=2, label='Regression line')
plt.xlabel('X values')
plt.ylabel('y values')
plt.title('Linear Regression using `sklearn`')
plt.legend()
plt.show()
#make predictions using the test set
X_test=np.asarray([10,15,20,25,30])
X_test=X_test.reshape(-1,1)
print("Test Set:\n", X_test)
yhat_test=model.predict(X_test)
print("Predictions:\n", yhat_test)
```
