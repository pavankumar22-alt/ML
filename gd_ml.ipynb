```python
import numpy as np
```

```python
x=np.asarray([20,15,10,13,12])
y=np.asarray([44,22,25,32,27])
```

```python
b1=0
b0=0
epochs=1000
l=0.001
n=np.size(x)
```

```python
for i in range(epochs):
  y_pre=b1*x+b0
  derv_b1=(-2/n)*sum(x*(y-y_pre))
  derv_b0=(-2/n)*sum(y-y_pre)
  b1=b1-l*derv_b1
  b0=b0-l*derv_b0
print(b1,b0)
```
